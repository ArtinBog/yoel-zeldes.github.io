```python
import numpy as np
from scipy.stats import kendalltau
import seaborn as sns
from matplotlib import cm
import matplotlib
from matplotlib import pyplot as plt
import pylab
plt.rcParams['figure.figsize'] = (200.0, 10.0)
%matplotlib inline

figsize = (40, 20)
```

```python
rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -2*(x*.1)**4.5 + .5 * x + rs.normal(size=1000)
plt.figure(figsize=figsize)
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.hexbin(x, y, gridsize=(40, 15), cmap=cm.Blues_r, bins=2000)
```

```python
sns.set(style={'axes.facecolor': '#4863A0'})
plt.figure(figsize=figsize)
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
sns.distplot((rs.uniform(size=1000) + 1) ** 3, kde=False, bins=100, color='#6495ED')
```
